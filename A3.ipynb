# Implement an N-gram language model in Python to analyze a text sequence and predict the most probable next word.

```python
import nltk
from collections import Counter

nltk.download('punkt')

# 1. Corpus
corpus = "natural language processing is easy and language processing is useful"

# 2. Input n
n = int(input("Enter n for n-gram (>=1): "))

# 3. Tokenize
tokens = nltk.word_tokenize(corpus.lower())

if n == 1:
    # unigram: predict most frequent word
    counts = Counter(tokens)
    predicted_word = counts.most_common(1)[0][0]
else:
    # n-gram for n >= 2
    ngrams = list(zip(*[tokens[i:] for i in range(n)]))
    counts = Counter(ngrams)

    context_counts = Counter([ng[-2] for ng in ngrams])

    context = input("Enter context word: ").lower()

    # 5. Predict next word
    predicted_word = None
    best_prob = 0
    for ng, c in counts.items():
        if ng[-2] == context:
            p = c / context_counts[context]  # probability (used internally)
            if p > best_prob:
                predicted_word = ng[-1]
                best_prob = p

print("Predicted next word:", predicted_word)
```

```python
import nltk
from collections import Counter

nltk.download('punkt')

# 1. Corpus
corpus = "natural language processing is easy and language processing is useful"

# 2. Input n
n = int(input("Enter n for n-gram (>=1): "))

# 3. Tokenize
tokens = nltk.word_tokenize(corpus.lower())

vocab = set(tokens)  # vocabulary for Laplace smoothing

if n == 1:
    # Unigram: predict the most frequent word
    counts = Counter(tokens)
    predicted_word = counts.most_common(1)[0][0]
else:
    # n-gram for n >= 2
    ngrams = list(zip(*[tokens[i:] for i in range(n)]))
    counts = Counter(ngrams)

    # context counts (single-word context = last word before target)
    context_counts = Counter([ng[-2] for ng in ngrams])

    # 4. Input context (single word)
    context = input("Enter context word: ").lower()

    # 5. Predict next word with Laplace smoothing
    predicted_word = None
    best_prob = 0
    V = len(vocab)  # vocabulary size
    for word in vocab:
        count_ng = counts.get((context, word), 0)
        count_ctx = context_counts.get(context, 0)
        # Laplace smoothed probability
        p = (count_ng + 1) / (count_ctx + V)
        if p > best_prob:
            predicted_word = word
            best_prob = p

print("Predicted next word:", predicted_word)
```

```python
from sklearn.preprocessing import LabelEncoder
from keras.utils import to_categorical

# Single line
line = "Artificial intelligence (AI) is transforming the way we live and work."

# Step 1: Tokenize and clean
words = line.lower().replace('(', '').replace(')', '').replace('.', '').split()
unique_words = sorted(set(words))  # vocabulary for this line only

# Step 2: Encode words as integers
label_encoder = LabelEncoder()
integer_encoded = label_encoder.fit_transform(words)

# Step 3: One-hot encode
onehot_encoded = to_categorical(integer_encoded)

print("Vocabulary:", unique_words)
print("Integer encoding:", integer_encoded)
print("One-hot encoded vectors:\n", onehot_encoded)
```
