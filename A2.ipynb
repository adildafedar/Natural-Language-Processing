# Develop a Python script to perform part-of-speech tagging on a short text dataset.

```python
import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
```

```python
dataset = {
    "Natural language processing is easy.",
    "Python is a programming language.",
    "Machine learning is a subfield of artificial intelligence."
}
```

Part of Speech (POS) tagging refers to assigning each word of a sentence to its part of speech. It is significant as it helps to give a better syntactic overview of a sentence.

```python
for i in dataset:
  tokens = nltk.word_tokenize(i)
  tagged = nltk.pos_tag(tokens)
  print(f"{i}")
  for word, tag in tagged:
    print(f"{word} {tag}")
```
