Use LSTM networks with Tensor Flow or Py Torch in Python to perform sequence labeling on a
small text dataset.

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
```

```python
text = [
    "I love natural language processing",
    "I love machine learning",
    "machine learning is fun",
    "natural language processing is powerful"
]
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(text)

total_words = len(tokenizer.word_index) + 1
```

```python
input_sequences = []

for line in text:
    token_list = tokenizer.texts_to_sequences([line])[0]

    for i in range(1, len(token_list)):
        n_gram = token_list[:i+1]
        input_sequences.append(n_gram)
```

```python
max_seq_len = max(len(x) for x in input_sequences)

input_sequences = np.array(
    pad_sequences(input_sequences, maxlen=max_seq_len, padding='pre')
)
```

```python
X = input_sequences[:, :-1]
y = input_sequences[:, -1]

y = tf.keras.utils.to_categorical(y, num_classes=total_words)
```

```python
model = Sequential([
    Embedding(total_words, 50, input_length=max_seq_len-1),
    LSTM(100),
    Dense(total_words, activation='softmax')
])

model.compile(
    loss='categorical_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)

model.summary()
```

```python
model.fit(X, y, epochs=200, verbose=1)
```

```python
def predict_next_word(text):

    token_list = tokenizer.texts_to_sequences([text])[0]
    token_list = pad_sequences([token_list], maxlen=max_seq_len-1, padding='pre')

    predicted = model.predict(token_list, verbose=0)

    predicted_word = tokenizer.index_word[np.argmax(predicted)]

    return predicted_word
```

```python
print(predict_next_word("I love"))
```

## Conclusion :
LSTM neural networks were implemented using TensorFlow to perform sequence labeling and predict the next word from a small text dataset.
