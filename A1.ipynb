# Implement tokenization, stemming, and lemmatization on a small corpus of text.

```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
sentence = "Lorem ipsum dolor sit amet, consectetur adipiscing elit, sed do eiusmod tempor incididunt ut labore et dolore magna aliqua. Ut enim ad minim veniam, quis nostrud exercitation ullamco laboris nisi ut aliquip ex ea commodo consequat."
tokens = nltk.word_tokenize(sentence)
```

```python
tokens
```

Stemming is a process in which root word is extracted.

```python

from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
for token in tokens :
  stemmed_tokens = stemmer.stem(token)
  print(stemmed_tokens)
```

```python
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
for token in tokens:
  lemmatized_tokens = lemmatizer.lemmatize(token)
  print(lemmatized_tokens)
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
filtered_tokens = [token for token in tokens if token.lower() not in stop_words]
print(filtered_tokens)
```
